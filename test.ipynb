### Checking if Tickers in my portfolio are available in yfinance

```python
dat2 = yf.Ticker("IMEU.L")
dat2.history(period='1mo')
```

```python
STOCKS = ["IMEU.L", "MSFT", "IAAA.L", "SGLN.L", "VHYD.L"]

for stock in STOCKS:
    dat = yf.Ticker(stock)
    dat.history(period='1mo')
    dat.history(period='1mo').to_csv(f"{stock}.csv")
   
```
